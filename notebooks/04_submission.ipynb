# 04 - Submission

Load trained model bundle and processed test set, then generate `submission.csv`.

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().resolve().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

import pandas as pd
from src import config
from src.inference import predict_with_ensemble
from src.utils import load_pickle
```

```python
if not Path(config.MODEL_BUNDLE_PATH).exists():
    raise FileNotFoundError('Run notebooks/03_model_training.ipynb first.')
if not Path(config.MERGED_TEST_PATH).exists():
    raise FileNotFoundError('Run notebooks/02_feature_engineering.ipynb first.')

model_bundle = load_pickle(config.MODEL_BUNDLE_PATH)
merged_test = pd.read_pickle(config.MERGED_TEST_PATH)
```

```python
submission = predict_with_ensemble(model_bundle, merged_test)
submission.to_csv(config.SUBMISSION_PATH, index=False)
print(f'Created submission: {config.SUBMISSION_PATH}')
submission.head()
```
